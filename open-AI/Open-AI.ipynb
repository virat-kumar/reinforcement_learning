```python
# Import gym and make an environment of game acrobot
import gym
```

```python
# Make an environment
env = gym.make('Acrobot-v1')
```

```python
# Reset all the environment parameters
#The output is a numpy array of current evrironment state/Obsetvation 
env.reset()
```

```python
# Displays the current environment 
env.render()
```

```python
# create a action space sample 
env.action_space.sample()
```

```python
# obser
```

```python
# perform an action
# The step function returns 4 outputs 
# 1)Observation/new_state This the new observation we get after performing an action
# 2)reward This is the reward of the new action taken
# 3)done  Its a boolean value which tells the agent has successfully completed its work
# 4) Info  Some parameters we need from environment used for debugging
observation, reward, done, info=env.step(env.action_space.sample())
```

```python
# Lets make an environment and perform some action
env = gym.make('MountainCarContinuous-v0') # try for different environements
observation = env.reset()
for t in range(1000):
        env.render()
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        print (observation, reward, done, info)
        if done:
            print("Finished after {} timesteps".format(t+1))
            break
env.close()
```

```python
'''
Discrete is non-negative possible values, above 0 or 1 are equivalent 
        to left and right movement for CartPole balancing
        
Box means n-dim array

action-space & observation-space describes what is the valid format for that particular env to work on with.
'''
import gym
env = gym.make('CartPole-v0')
print(env.action_space) #[Output: ] Discrete(2) 
print(env.observation_space) # [Output: ] Box(4,)
env.close()
env = gym.make('MountainCarContinuous-v0')
print(env.action_space) #[Output: ] Box(1,)
print(env.observation_space) #[Output: ] Box(2,)
env.close()
```
